```python
3import sqlite3

con = sqlite3.connect('E_ticaret.db')

cursor = con.cursor()

cursor.execute('CREATE TABLE IF NOT EXISTS Uye (uye_id INT,kayit_id INT ,isim_soyisim TEXT,siparisg_id INT,adres_id INT,adres TEXT)')

for i in range(5):
  uye_id = int(input( 'uye_id girin'))
  kayit_id = int(input('kayıt numarası girin'))
  isim_soyisim = input('isim soy isim girin')
  siparisg_id = int(input('siparişg_id geçmişi giriniz'))
  adres_id = int(input('adres_id giriniz'))
  adres = input('adres giriniz.')
  cursor.execute('INSERT INTO Uye Values(?,?,?,?,?,?)',(uye_id,kayit_id,isim_soyisim,siparisg_id,adres_id,adres))

cursor.execute('CREATE TABLE IF NOT EXISTS Kayit (kayit_id İNT ,email TEXT,sifre TEXT,kullanici_adi TEXT)')
for i in range(5):
  kayit_id = int(input('kayıt numarası giirn'))
  kullanici_adi = input('kullanıcı adı girin')
  email = (input('email giriniz'))
  sifre = input('sifre giriniz')
  cursor.execute('Insert into Kayit Values(?,?,?,?)',(kayit_id,kullanici_adi,email,sifre))

cursor.execute('CREATE TABLE IF NOT EXISTS Siparis (siparis_id İNT ,odemeyontemi TEXT,sepet_id İNT,uye_id İNT,adres_id İNT,uye_fatura_id İNT,tarih TEXT,siparis_durum_id İNT)')
for i in range(5):
  siparis_id = int(input('kayıt numarası giirn'))
  odemeyontemi = input('isim soy isim girin')
  sepet_id = int(input('siparişg_id geçmişi giriniz'))
  uye_id = int(input('uye_id giriniz'))
  adres_id = int(input('adres id giriniz.'))
  uye_fatura_id = int(input('fatura id giriniz.'))
  tarih = input('tarih giriniz.')
  siparis_durum_id = int(input('sipariş durum id si girin.'))
  cursor.execute('Insert into Siparis Values(?,?,?,?,?,?,?,?)',(siparis_id,odemeyontemi,sepet_id,uye_id,adres_id,uye_fatura_id,tarih,siparis_durum_id))


cursor.execute('CREATE TABLE IF NOT EXISTS Urunler (urun_id İNT,urun_kod İNT ,urun_adi TEXT,kategori_id İNT,aciklama TEXT,fiyat İNT)')
for i in range(5):
  urun_id = int(input('urun_id si girin'))
  urun_kod = int(input('urun_kodu girin'))
  urun_adi = input('ürün adı giriniz')
  kategori_id = int(input('kategori_id giriniz'))
  aciklama = input('ürün açıklaması girin.')
  fiyat = int(input('ürün fiyati girin'))
  cursor.execute('Insert into Urunler Values(?,?,?,?,?,?)',(urun_id,urun_kod,urun_adi,kategori_id,aciklama,fiyat))


cursor.execute('CREATE TABLE IF NOT EXISTS uye_fatura (uye_fatura_id İNT,uye_id İNT ,faturabilgi TEXT,adres_id İNT)')
for i in range(5):
  uye_fatura_id = int(input('fatura_id si giriniz.'))
  uye_id = int(input('uye_id si girin'))
  faturabilgi = input('faturabilgisi  girin')
  adres_id = int(input('adres_id si  giriniz'))
  cursor.execute('Insert into uye_fatura Values(?,?,?,?)',(uye_fatura_id,uye_id,faturabilgi,adres_id))


cursor.execute('CREATE TABLE IF NOT EXISTS uye_adres (uye_adres_id İNT,uye_id İNT ,adresbasligi TEXT,adres TEXT)')
for i in range(5):
  uye_adres_id = int(input('uye_adres_id si girin'))
  uye_id = int(input('uye_id si girin'))
  adresbasligi = input('adresbaşlığı  girin')
  adres = input('adres  giriniz')
  cursor.execute('Insert into uye_adres Values(?,?,?,?)',(uye_adres_id,uye_id,adresbasligi,adres))


cursor.execute('CREATE TABLE IF NOT EXISTS uye_yorum (uye_yorum_id İNT,uye_id İNT ,urun_id İNT,yorum TEXT)')
for i in range(5):
  uye_yorum_id = int(input('uye_yorum_id si girin'))
  uye_id = int(input('uye_id si girin'))
  urun_id = int(input('urun_id  girin'))
  yorum = input('yorum  giriniz')
  cursor.execute('Insert into uye_yorum Values(?,?,?,?)',(uye_yorum_id,uye_id,urun_id,yorum))


cursor.execute('CREATE TABLE IF NOT EXISTS arama_gecmisi (arama_gecmisi_id İNT,uye_id İNT ,aratilan_kelime TEXT)')
for i in range(5):
  arama_gecmisi_id = int(input('arama_gecmisi_id si girin'))
  uye_id = int(input('uye_id si girin'))
  aratilan_kelime = input('aratilan kelimeyi girin')
  cursor.execute('Insert into arama_gecmisi Values(?,?,?)',(arama_gecmisi_id,uye_id,aratilan_kelime))


con.commit()


con.close()
```
